# ğŸ“1. Student Report Card Generator

## ğŸ¯ Objective
This program collects student information, accepts subject marks with proper validation, calculates the total and average marks, assigns a grade, and finally prints and saves a report card.

---

## ğŸ› ï¸ Features
- ğŸ“¥ Takes student name and age as input
- ğŸ“š Accepts marks for **Math**, **Science**, and **English**
- âœ… Validates marks to ensure they're between 0 and 100
- â• Calculates total and average marks
- ğŸ“ Assigns grades based on average:
  - A+ (90 and above)
  - A (75â€“89)
  - B (60â€“74)
  - C (40â€“59)
  - F (below 40)
- ğŸ–¨ï¸ Prints a formatted report card
- ğŸ’¾ Saves the report card to a `.txt` file
- ğŸ” Uses exception handling for user input and file errors

---

## ğŸ§  Python Concepts Used
- **Input Handling** using `input()`
- **Data Types**: `int`, `float`, and `str`
- **Functions** for modular design
- **Conditionals** (`if-elif-else`) for grading logic
- **Loops** and **exception handling** for mark validation
- **File Handling** to write report card to `report_card.txt`

---

## ğŸ–¥ï¸ Sample Output

Enter student name: Sakthi Enter student age: 20 Enter marks in Math: 85 Enter marks in Science: 91 Enter marks in English: 78

====== STUDENT REPORT CARD ====== Name : Sakthi Age : 20 Math : 85.0 Science: 91.0 English: 78.0 Total : 254.0 Average: 84.67 Grade : A

```python
# ğŸ¯ Collect Student Information
student_name = input("Enter student name: ")
student_age = int(input("Enter student age: "))

# ğŸ“¥ Take Subject Marks with Validation
def get_mark(subject):
    while True:
        try:
            mark = float(input(f"Enter marks in {subject}: "))
            if 0 <= mark <= 100:
                return mark
            else:
                print("âŒ Please enter a value between 0 and 100.")
        except ValueError:
            print("âŒ Invalid input. Please enter a number.")

math = get_mark("Math")
science = get_mark("Science")
english = get_mark("English")

# â— Calculate Total & Average
total_marks = math + science + english
average_marks = total_marks / 3

# ğŸ“ Grade Assignment Logic
def assign_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 75:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 40:
        return "C"
    else:
        return "F"

grade = assign_grade(average_marks)

# ğŸ–¨ï¸ Print Report Card
def print_report():
    print("\n====== STUDENT REPORT CARD ======")
    print(f"Name   : {student_name}")
    print(f"Age    : {student_age}")
    print(f"Math   : {math}")
    print(f"Science: {science}")
    print(f"English: {english}")
    print(f"Total  : {total_marks}")
    print(f"Average: {average_marks:.2f}")
    print(f"Grade  : {grade}")
    print("==================================")

print_report()

# ğŸ’¾ Save Report Card to File with Exception Handling
try:
    with open("report_card.txt", "w") as file:
        file.write("====== STUDENT REPORT CARD ======\n")
        file.write(f"Name   : {student_name}\n")
        file.write(f"Age    : {student_age}\n")
        file.write(f"Math   : {math}\n")
        file.write(f"Science: {science}\n")
        file.write(f"English: {english}\n")
        file.write(f"Total  : {total_marks}\n")
        file.write(f"Average: {average_marks:.2f}\n")
        file.write(f"Grade  : {grade}\n")
        file.write("==================================\n")
    print("\nâœ… Report card saved successfully to 'report_card.txt'")
except Exception as e:
    print("âŒ Error saving file:", e)
```

## ğŸ¯ Objective
This program simulates a basic ATM system in the command line, allowing the user to:

---

## ğŸ› ï¸ Features

- ğŸ’¸ Withdrawal with balance check

## ğŸ§  Python Concepts Used

## ğŸ–¥ï¸ Sample Output

===== ğŸ§ ATM Menu =====

ğŸ§¾ Your current balance is â‚¹1000

```python
# ğŸ¦ Simple ATM Simulator

# âœ… Step 1: Initialize Balance
balance = 1000  # Starting balance in â‚¹

# âœ… Step 2: Define ATM Functions
def check_balance():
    print(f"\nğŸ§¾ Your current balance is â‚¹{balance}")

def deposit(amount):
    global balance
    balance += amount
    print(f"âœ… â‚¹{amount} deposited successfully. New balance: â‚¹{balance}")

def withdraw(amount):
    global balance
    if amount > balance:
        print("âŒ Insufficient balance. Transaction cancelled.")
    else:
        balance -= amount
        print(f"âœ… â‚¹{amount} withdrawn successfully. Remaining balance: â‚¹{balance}")

# âœ… Step 3: Run the ATM Menu in a Loop
while True:
    print("\n===== ğŸ§ ATM Menu =====")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    try:
        choice = int(input("Enter your choice (1â€“4): "))
    except ValueError:
        print("âŒ Invalid input. Please enter a number from 1 to 4.")
        continue

    # âœ… Step 4: Handle User Selection
    if choice == 1:
        check_balance()
    elif choice == 2:
        try:
            amount = float(input("Enter amount to deposit: â‚¹"))
            if amount <= 0:
                print("âŒ Deposit amount must be greater than 0.")
            else:
                deposit(amount)
        except ValueError:
            print("âŒ Invalid input. Please enter a valid number.")
    elif choice == 3:
        try:
            amount = float(input("Enter amount to withdraw: â‚¹"))
            if amount <= 0:
                print("âŒ Withdrawal amount must be greater than 0.")
            else:
                withdraw(amount)
        except ValueError:
            print("âŒ Invalid input. Please enter a valid number.")
    elif choice == 4:
        print("ğŸ‘‹ Thank you for using the ATM. Have a great day!")
        break
    else:
        print("âŒ Invalid choice. Please select between 1 and 4.")
```

# ğŸ§‘â€ğŸ“3. Student Record Manager

## ğŸ¯ Objective
This Python program manages a student database where users can:
- âœ… Add a new student
- âœ… View a specific student's record
- âœ… Update marks for a subject
- âœ… View all student records
- âœ… Delete a student
- âœ… Search for students by partial name

---

## ğŸ› ï¸ Features
- ğŸ“‚ Stores student data persistently using a JSON file
- ğŸ§  Calculates average and assigns grade (A+, A, B, C, F)
- âœï¸ Allows subject-wise mark updates
- ğŸ” Supports partial name search
- âŒ Graceful handling of invalid input and duplicate entries

---

## ğŸ§  Python Concepts Used
- **File handling** with `json` and `os` modules
- **Functions** for structured logic (`add_student()`, `view_student()`, etc.)
- **Dictionaries** to store student data
- **Conditionals and loops** for menu control
- **Input validation** using `try-except`

---

## ğŸ–¥ï¸ Sample Output

=== STUDENT DATABASE MENU ===

Add Student

View Student

Update Marks

View All Students

Delete Student

Search by Partial Name

Exit

Enter your choice (1-7): 1 Enter student name: Sakthivelan Enter age: 20 Enter Math marks: 90 Enter Science marks: 85 Enter English marks: 88 âœ… Sakthivelan's record added successfully.

```python
import json
import os

# File to store student records
DATA_FILE = "students.json"
students = {}

# âœ… Load existing data from file if available
def load_data():
    global students
    if os.path.exists(DATA_FILE):
        with open(DATA_FILE, "r") as f:
            students = json.load(f)

# âœ… Save current data to file
def save_data():
    with open(DATA_FILE, "w") as f:
        json.dump(students, f, indent=4)

# âœ… Calculate average and grade
def calculate_grade(avg):
    if avg >= 90:
        return 'A+'
    elif avg >= 75:
        return 'A'
    elif avg >= 60:
        return 'B'
    elif avg >= 50:
        return 'C'
    else:
        return 'F'

# âœ… Add Student
def add_student():
    name = input("Enter student name: ").title()
    if name in students:
        print("âš ï¸ Student already exists.")
        return
    try:
        age = int(input("Enter age: "))
        marks = {
            "Math": float(input("Enter Math marks: ")),
            "Science": float(input("Enter Science marks: ")),
            "English": float(input("Enter English marks: "))
        }
        avg = sum(marks.values()) / len(marks)
        grade = calculate_grade(avg)
        students[name] = {
            "age": age,
            "marks": marks,
            "average": avg,
            "grade": grade
        }
        save_data()
        print(f"âœ… {name}'s record added successfully.")
    except ValueError:
        print("âŒ Invalid input. Please enter correct data types.")

# âœ… View Student by Name
def view_student():
    name = input("Enter student name to search: ").title()
    if name in students:
        record = students[name]
        print(f"\nğŸ“„ Record for {name}")
        print(f"Age: {record['age']}")
        for subject, score in record["marks"].items():
            print(f"{subject}: {score}")
        print(f"Average: {record['average']:.2f}")
        print(f"Grade: {record['grade']}")
    else:
        print("âŒ Student not found.")

# âœ… Update Marks
def update_marks():
    name = input("Enter student name to update: ").title()
    if name in students:
        subject = input("Enter subject to update (Math/Science/English): ").title()
        if subject in students[name]["marks"]:
            try:
                new_mark = float(input(f"Enter new marks for {subject}: "))
                students[name]["marks"][subject] = new_mark
                avg = sum(students[name]["marks"].values()) / 3
                students[name]["average"] = avg
                students[name]["grade"] = calculate_grade(avg)
                save_data()
                print("âœ… Marks updated successfully.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        else:
            print("âŒ Invalid subject.")
    else:
        print("âŒ Student not found.")

# âœ… View All Students
def view_all():
    if not students:
        print("âš ï¸ No records available.")
        return
    print("\nğŸ“‹ All Student Records")
    for name, info in students.items():
        print(f"\nğŸ”¸ Name: {name}, Age: {info['age']}")
        for subject, score in info["marks"].items():
            print(f"  {subject}: {score}")
        print(f"  Average: {info['average']:.2f}")
        print(f"  Grade: {info['grade']}")

# âœ… Delete Student Record
def delete_student():
    name = input("Enter student name to delete: ").title()
    if name in students:
        del students[name]
        save_data()
        print(f"ğŸ—‘ï¸ {name}'s record deleted.")
    else:
        print("âŒ Student not found.")

# âœ… Search by Partial Name
def search_partial():
    keyword = input("Enter part of the student name: ").title()
    found = False
    for name in students:
        if keyword in name:
            found = True
            print(f"\nğŸ” Match: {name}")
            record = students[name]
            print(f"  Age: {record['age']}")
            for subject, score in record["marks"].items():
                print(f"    {subject}: {score}")
            print(f"  Average: {record['average']:.2f}, Grade: {record['grade']}")
    if not found:
        print("âŒ No matches found.")

# âœ… Main Menu
def main():
    load_data()
    while True:
        print("\n=== STUDENT DATABASE MENU ===")
        print("1. Add Student")
        print("2. View Student")
        print("3. Update Marks")
        print("4. View All Students")
        print("5. Delete Student")
        print("6. Search by Partial Name")
        print("7. Exit")

        try:
            choice = int(input("Enter your choice (1-7): "))
        except ValueError:
            print("âŒ Invalid choice. Enter a number from 1-7.")
            continue

        if choice == 1:
            add_student()
        elif choice == 2:
            view_student()
        elif choice == 3:
            update_marks()
        elif choice == 4:
            view_all()
        elif choice == 5:
            delete_student()
        elif choice == 6:
            search_partial()
        elif choice == 7:
            print("ğŸ‘‹ Exiting Student Database. Goodbye!")
            break
        else:
            print("âŒ Invalid choice.")

# âœ… Entry Point
if __name__ == "__main__":
    main()
```

# ğŸ›’4. Shopping Cart System (CLI Version)

## ğŸ¯ Objective
This program simulates a basic **shopping cart system** in the command line where users can:
- âœ… View available products
- âœ… Add products to a cart
- âœ… Remove products from the cart
- âœ… View cart summary
- âœ… Checkout with tax calculation
- âœ… Exit the application

---

## ğŸ› ï¸ Features
- ğŸ“¦ Predefined product catalog with prices
- ğŸ›ï¸ Interactive menu for seamless user experience
- ğŸ§º Dynamic cart with quantity tracking
- ğŸ“‰ Product removal functionality
- ğŸ’³ Checkout with 10% GST and total calculation
- âŒ Handles invalid inputs and edge cases

---

## ğŸ§  Python Concepts Used
- **Dictionaries** for product catalog and cart
- **Functions** for organized, reusable logic (`add_to_cart()`, `view_cart()`, etc.)
- **Loops and conditionals** to control menu and logic flow
- **Exception handling** for robust user input
- **String manipulation** and formatting

---

## ğŸ–¥ï¸ Sample Output

=== SHOPPING MENU ===

View Products

Add to Cart

Remove from Cart

View Cart

Checkout

Exit Enter your choice (1-6): 1

ğŸ›ï¸ Available Products: Apple: â‚¹20 Banana: â‚¹10 Milk: â‚¹45 Bread: â‚¹30 Eggs: â‚¹60

Enter your choice (1-6): 2 Enter product name to add: milk Enter quantity for milk: 2 âœ… 2 x milk added to cart.

Enter your choice (1-6): 4

ğŸ§¾ CART SUMMARY: Milk - â‚¹45 x 2 = â‚¹90 ğŸ§® Total: â‚¹90

ğŸ’³ CHECKOUT BILL: Milk - â‚¹45 x 2 = â‚¹90

Subtotal: â‚¹90 GST (10%): â‚¹9.0 ğŸ’° Grand Total: â‚¹99.0 ğŸ§¾ Thank you for shopping with us!

```python
# ğŸ›’ SHOPPING CART SYSTEM (CLI Version)

# âœ… Product catalog (Dictionary)
products = {
    "apple": 20,
    "banana": 10,
    "milk": 45,
    "bread": 30,
    "eggs": 60
}

# âœ… Empty shopping cart
cart = {}

# âœ… Add product to cart
def add_to_cart():
    item = input("Enter product name to add: ").lower()
    if item in products:
        try:
            qty = int(input(f"Enter quantity for {item}: "))
            if qty <= 0:
                print("âŒ Quantity must be positive.")
                return
            if item in cart:
                cart[item] += qty
            else:
                cart[item] = qty
            print(f"âœ… {qty} x {item} added to cart.")
        except ValueError:
            print("âŒ Invalid quantity.")
    else:
        print("âŒ Product not found.")

# âœ… Remove product from cart
def remove_from_cart():
    item = input("Enter product name to remove: ").lower()
    if item in cart:
        del cart[item]
        print(f"ğŸ—‘ï¸ {item} removed from cart.")
    else:
        print("âŒ Item not in cart.")

# âœ… View cart summary
def view_cart():
    if not cart:
        print("ğŸ›’ Your cart is empty.")
        return
    print("\nğŸ§¾ CART SUMMARY:")
    total = 0
    for item, qty in cart.items():
        price = products[item]
        subtotal = price * qty
        print(f"{item.title()} - â‚¹{price} x {qty} = â‚¹{subtotal}")
        total += subtotal
    print(f"ğŸ§® Total: â‚¹{total}")

# âœ… Checkout and print bill
def checkout():
    if not cart:
        print("âŒ Cart is empty. Cannot checkout.")
        return
    print("\nğŸ’³ CHECKOUT BILL:")
    total = 0
    for item, qty in cart.items():
        price = products[item]
        subtotal = price * qty
        print(f"{item.title()} - â‚¹{price} x {qty} = â‚¹{subtotal}")
        total += subtotal
    tax = round(total * 0.1, 2)
    grand_total = total + tax
    print(f"\nSubtotal: â‚¹{total}")
    print(f"GST (10%): â‚¹{tax}")
    print(f"ğŸ’° Grand Total: â‚¹{grand_total}")
    print("ğŸ§¾ Thank you for shopping with us!")
    cart.clear()

# âœ… Menu loop
def main():
    while True:
        print("\n=== SHOPPING MENU ===")
        print("1. View Products")
        print("2. Add to Cart")
        print("3. Remove from Cart")
        print("4. View Cart")
        print("5. Checkout")
        print("6. Exit")

        try:
            choice = int(input("Enter your choice (1-6): "))
        except ValueError:
            print("âŒ Please enter a valid number.")
            continue

        if choice == 1:
            print("\nğŸ›ï¸ Available Products:")
            for item, price in products.items():
                print(f"{item.title()}: â‚¹{price}")
        elif choice == 2:
            add_to_cart()
        elif choice == 3:
            remove_from_cart()
        elif choice == 4:
            view_cart()
        elif choice == 5:
            checkout()
        elif choice == 6:
            print("ğŸ‘‹ Thank you! Visit again.")
            break
        else:
            print("âŒ Invalid choice.")

# âœ… Entry Point
if __name__ == "__main__":
    main()
```

## ğŸ–¥ï¸ Sample Output

ğŸ¯ Welcome to the Python Quiz! ğŸ”¢ Choose the correct option (A, B, C, or D)

```python
# ğŸ§  MINI QUIZ APPLICATION (COLAB VERSION)

# âœ… Question Bank: List of dictionaries
questions = [
    {
        "question": "What is the capital of France?",
        "options": ["A. Paris", "B. Rome", "C. Berlin", "D. Madrid"],
        "answer": "A"
    },
    {
        "question": "Which data type is immutable in Python?",
        "options": ["A. List", "B. Dictionary", "C. Tuple", "D. Set"],
        "answer": "C"
    },
    {
        "question": "What does HTML stand for?",
        "options": ["A. Hyper Trainer Markup Language", "B. HyperText Markup Language", "C. HighText Machine Language", "D. None of the above"],
        "answer": "B"
    },
    {
        "question": "Which company developed Python?",
        "options": ["A. Microsoft", "B. Apple", "C. Google", "D. None of the above"],
        "answer": "D"
    },
    {
        "question": "Which keyword is used for function in Python?",
        "options": ["A. func", "B. define", "C. def", "D. function"],
        "answer": "C"
    }
]

# âœ… Function to play quiz
def play_quiz():
    score = 0
    print("\nğŸ¯ Welcome to the Python Quiz!")
    print("ğŸ”¢ Choose the correct option (A, B, C, or D)\n")

    for idx, q in enumerate(questions, start=1):
        print(f"Q{idx}: {q['question']}")
        for opt in q["options"]:
            print(opt)
        try:
            user_answer = input("Your answer: ").strip().upper()
            if user_answer == q["answer"]:
                print("âœ… Correct!\n")
                score += 1
            else:
                print(f"âŒ Incorrect. The right answer is {q['answer']}.\n")
        except Exception as e:
            print("âŒ An error occurred:", e)

    # âœ… Final result
    print(f"ğŸ§¾ You scored {score} out of {len(questions)}")

    if score == len(questions):
        print("ğŸ‰ Excellent! You're a Python Pro!")
    elif score >= len(questions) * 0.6:
        print("ğŸ‘ Good job! Keep practicing.")
    else:
        print("ğŸ“š Keep learning. You can do better!")

# âœ… Main loop
def main():
    while True:
        play_quiz()
        replay = input("\nğŸ” Do you want to try again? (yes/no): ").strip().lower()
        if replay != "yes":
            print("ğŸ‘‹ Thanks for playing the quiz. Goodbye!")
            break

# âœ… Entry Point
if __name__ == "__main__":
    main()
```
